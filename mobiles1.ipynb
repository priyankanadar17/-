The data we use is usually split into training data and test data. The training set contains a known output and the model learns on this data in order to be generalized to other data later. We have the test dataset in order to test our model’s prediction.

Accuracy is a performance measure and it is a ratio of correctly predicted observation to the total observations. One may think that, if we have high accuracy then our model is best. Yes, accuracy is a great measure but only when you have symmetric datasets where values of false positive and false negatives are almost same. Therefore, you have to look at other parameters to evaluate the performance of your model.

```python
rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
```

Confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known. It allows the visualization of the performance of an algorithm.
